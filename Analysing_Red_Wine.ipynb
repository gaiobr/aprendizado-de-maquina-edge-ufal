### **Distribuição de propabilidade**

```
# Carregar biblioteca
import seaborn as sns
```

```
# Distribuição de probabilidade comparando o teor alcoólico por nível de qualidade
for x in quality:
  sns.distplot(df[df['quality'] == x].alcohol, hist = False, rug = False, 
             label = x) 

plt.title('Teor Alcoólico por Qualidade')
plt.show()

# Distribuição de probabilidade comparando a acidez volátil por nível de qualidade
for x in quality:
  sns.distplot(df[df['quality'] == x].volatile_acidity, hist = False, rug = False, 
             label = x) 

plt.title('Acidez Volátil por Qualidade')
plt.show()

```

Demonstrando a tendência da acidez volátil e teor alcoólico para os vinhos de maior qualidade

# **Análise Estatística dos Dados**

### **Estudo de correlação entre variáveis**

```
# Calcular correlação entre qualidade e os demais atributos
for col in df.columns[:-1]:
  print(col, ': ', round(df.quality.corr(df[col]),2))
```

```
# Calcular correlação entre os atributos
corr = df.corr()
corr
```

```
# Plotar coeficientes de correlação em um mapa de calor
sns.heatmap(corr, vmin = -1, vmax = 1, center = 0,
            cmap = sns.diverging_palette(20, 220, n = 200))
```

* As variáveis não demonstram correlação significativa com a variável qualidade.  
* As variáveis pH, densidade e ácido cítrico apresentam correlação moderada com acidez fixa 


# **Re-escalando os dados**

### **Usando média e desvio-padrão**

```
df_mod = df
df_mod.fixed_acidity = (df_mod.fixed_acidity - df_mod.fixed_acidity.mean())/df_mod.fixed_acidity.std()
df_mod.volatile_acidity = (df_mod.volatile_acidity - df_mod.volatile_acidity.mean())/df_mod.volatile_acidity.std()
df_mod.citric_acid = (df_mod.citric_acid - df_mod.citric_acid.mean())/df_mod.citric_acid.std()
df_mod.residual_sugar = (df_mod.residual_sugar - df_mod.residual_sugar.mean())/df_mod.residual_sugar.std()
df_mod.chlorides = (df_mod.chlorides - df_mod.chlorides.mean())/df_mod.chlorides.std()
df_mod.free_sulfur_dioxide = (df_mod.free_sulfur_dioxide - df_mod.free_sulfur_dioxide.mean())/df_mod.free_sulfur_dioxide.std()
df_mod.total_sulfur_dioxide = (df_mod.total_sulfur_dioxide - df_mod.total_sulfur_dioxide.mean())/df_mod.total_sulfur_dioxide.std()
df_mod.density = (df_mod.density - df_mod.density.mean())/df_mod.density.std()
df_mod.pH = (df_mod.pH - df_mod.pH.mean())/df_mod.pH.std()
df_mod.sulphates = (df_mod.sulphates - df_mod.sulphates.mean())/df_mod.sulphates.std()
df_mod.alcohol = (df_mod.alcohol - df_mod.alcohol.mean())/df_mod.alcohol.std()
```

```
# Calcular correlação entre os atributos
corr_mod = df_mod.corr()
corr_mod

```

```
# Plotar coeficientes de correlação em um mapa de calor
sns.heatmap(corr_mod, vmin = -1, vmax = 1, center = 0,
            cmap = sns.diverging_palette(20, 220, n = 200))
```

Mesmo após normalizar os dados, as correlações não se alteraram
